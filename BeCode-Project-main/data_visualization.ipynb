```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame()
df = pd.read_csv(r"C:\Users\Augustin\Downloads\BeCode-Project-main\data_a.csv")

#Q1 : "What's the relationship between a furnished apartment to rent in Brussels on its price versus an unfurnished apartment ?"

df.dropna(subset = ["To rent", "type", "zipcode", "Furnished", "Price", "Living Area"])
df = df[df['type'] == 'Appartement']
df = df[df['To rent'] == True]
df = df[(df["zipcode"] > 999) & (df["zipcode"] < 1300)]
df['Price/Living Area'] = df['Price']/df['Living Area']

sns.barplot(data = df, x = "Furnished", y = "Price/Living Area")

plt.show()

#Q2 : "How does having a garden in a furnished house affect the price when you want to buy a house in the center of Antwerp ?"

df = pd.DataFrame()
df = pd.read_csv(r"C:\Users\Augustin\Downloads\BeCode-Project-main\data_a.csv")

df.dropna(subset = ["To sell", "type", "zipcode", "Living Area", "Price", "Garden"])
df = df[df['type'] == 'Maison']
df = df[df['To sell'] == True]

# Calculate the median price of the living area of a house to buy in the center of Antwerp

df['Price/Living Area'] = df['Price']/df['Living Area']

# Calculate the effect that a garden has on that number. 

df = df[(df["zipcode"] > 1999) & (df["zipcode"] < 2190)]

sns.barplot(data = df, x = "Garden", y = "Price/Living Area")

plt.show()
```
